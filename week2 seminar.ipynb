Complexity
- A recap
    - what is required to run the algorithm
    - input is sufficiently large
        - aka input cannot be empty or small
- Time
    - Best, scenario when it is the algorithm is the fastest
        - binary search = O(1) when the item you are searching for is in the middle
        - my list is large, n-items
    - Worst, scenario when it is the algorithm is the slowest
        - binary search = O(log n) when 1st item (leftmost), last item (rightmost), or not found
    - Big-O, O(N) upper bound
        - how it scaling or order of magnitude
            - linear O(N)
            - exponential O(2^N)
        - ignore any constant
            - O(N) + O(N log N)
                - N + N log N = N (1 + log N) = N log N
            - N * (N - 1) = O(N^2)
        - only look at the biggest magnitude, order, exponent
    - Big-Theta, Î˜(N) tightest bound
        - view it as big-O but PRECISE aka tightest bound, due to the lower bound as well
- Space = input + aux
    - Input, cant control
        - bubble sort accepts a list of N items
            - input space complexity = O(N)
            - pass by reference, still consider it as the same input space
    - Auxiliary, extra space
        - bubble sort accepts a list of N items
            - O(1) aux space, because it only uses index (pointers), temp var for swapping
            - O(1) STILL consume space. It means it uses CONSTANT space (aka doesn't scale with input).
        - INPLACE = O(1) aux
    - Recursion?
        - will never be inplace aka wont have O(1) aux space
        - anything recursion, have to + O(recursive depth)
        - recursive stack = space used...
    - Examples:
        - Is bubble sort inplace? Yes, O(1) aux
        - Is merge sort inplace? No, resurive stack = O(log N) depth, O(N) aux for temp list merging
- Time vs Space
    - time complexity >= aux space complexity
        - time = O(N), aux space = O(N) is correct
        - time = O(N^2), aux space = O(1) is correct
        - time = O(N), aux space = O(N log N) is false
        - reason: you needd time to create the extra space and fill it up
