        - Only positive weight.

    - single source single sink?

- how to know the flow of the network?
    - sum of flow from the source
    - sum of flow into the sink / target / destination
- 2 main concepts
    - flow conservation
        - incoming == outgoing for any vertex EXCEPT source and sink
    - flow constraint
        - flow <= capacity for every edge

Residual Network
- Also known as potential network

    - What is it for? How much more flow you can pass through
- Backward/ reverse edge? 
    - Actual flow that has been passed through

Ford-Fulkerson
- Traversal
    - From: source / super source
    - To: sink / super sink
- Path augmentation
    - use BFS or DFS
        - but might need to modify
        - which is better? in general BFS
    - add flow from source to target
        - based on the minimum edge weight along the path in the residual network
        - have the edges with references to each other for O(1) access and update
- complexity O(F E) ~ O(V E^2)
    - ^ I give this here for you to start reasoning it out
        - F is the max flow
            - because you need to run it F times
            - flow increment each time
                - minimum increment is 1
        - E is the complexity of running BFS/DFS
            - because you need to run it multiple times, each time is O(V+E)
            - each time you need to find a path
                - path length is at least 1
