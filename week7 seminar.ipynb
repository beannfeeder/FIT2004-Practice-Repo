Dynamic Programming
- Brute force
    - in smarter way?
        - why? you break a problem down in a way that is repetitive (patterns!)
        - is it faster??? yes
        - what is the trade off??? memorize solution, do not recompute (thus faster), need memory
- What is it?
    - Overlapping subproblems
        - reuse partial solution
    - Optimal solution (or rather the subsolution)
        - to avoid recomputing because answer is wrong / not optimal
        - remember, the solution are reused, so it needs to be optimal or else everything is not optimal
    - vs Divide and Conquer?
        - DnC doesnt overlap or it is not solution is recomputed
        - mergesort
- Time vs Space
    - use space to reduce time (aka no recomputation)
- Recurrence relation?
    - Previous
        - used to find the time complexity of recursive functions...
    - but dynamic programming is difference
        - show how to find the optimal solution
            - at each memo
            - at each step
            - ... these represent the subproblems
                - and finally the overall problem
    - if you know the recurrence relation, programming is just based on the recurrence relation
    - 3 cases
        - base
            - initial value u initialize
            - requirement?
                - base case only have ONE answer / option no matter what
                - that ONE answer / option is always optimal, NO DECISION
            - IMPORTANT TO HAVE
                - why? this is where backtracking end
        - general case
            - depends on an earlier smaller case aka your subproblem(s)
                - subproblem(s) already have an optimal solution(s)
                - AND you need to make decisions which to use / take.
        - edge cases
            - like your general case
            - could reuse earlier subproblem
            - often no more decision needed
            - think of the maze question(s) from the tutorial when you are at the edge of the maze...
- aux space
    - memoization
        - 1D array = O(N)
        - 2D array = O(N*M)
        - 3D array = O(A*B*C)
    - this sets the lower bound for time complexity
        - need time to create the memo
        - <almost> every position / cell need to be filled
            - this isn't just O(1), it can be >O(1) like coin change, we go through each coin at each value
- time complexity
- 2 ways implement
    - top-down
        - go from large to small
        - need recursion
    - bottom-up
        - start from base case to large
        - do not need recursion
    - both have the same recurrence relation
    - which is more efficient? depends on the problem
        - if you can skip some subproblems, top-down
        - if you need all subproblems, bottom-up
- Finding optimal?
    - look at the recurrence relation
    - watch my videos for the process... EZ
- Backtracking optimal?
    - do not store complete solution at each memo
        - because it is a waste of space
    - only save the breadcrumbs
        - dijkstra, this is your prev
        - knapsack, this is your last item added
        - coin change, this is the last coin added
        - ... and so on and so on...
    - is it possible to not use memo for breadcrumbs?
        - YES, because there are concepts such as inclusion-exclusion principle
            - we saw this in lecture + tutorial
            - but not always
    - dont worry, even if u use memo for breadcrumbs, if within complexity you are GUCCI!
