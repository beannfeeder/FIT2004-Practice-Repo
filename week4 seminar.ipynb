QuickSort
- What is the invariants (at the end of the loop)? Assume increasing order...
    - left partition is <= pivot
    - right partition is > pivot
    - pivot is at the correct / final / sorted position 
- Partition with pivot
    - Why it matters? bcoz of the invariants above
    - Type of partitions?
        - Out of place
            - partition in a temporary list
            - it is not a good approach despite can be stable, because space
        - Lomuto
            - FIT1008
            - unstable
            - a lot of swapping, some items can be swapped more than once
        - Hoare's
            - less swapping
            - L and R, similar to what i shown with out of place but you do it inplace

            (left and right pointer overlap, pivot is at right postition)
    - Complexity for partition?
        - Time: O(N) because you need to go through all items, to compare vs pivot

    - Best: O(N log N). Why???

    - Average? O(N log N)... go watch the video

QuickSelect(mylist, k)
k is where u want the pivot to be
- What is it?
    - partition the list into 2 parts
    - based to pivot, pivot is given by the user as the input (aka the position i want)
        - we call this the k-order or k-value that is argument to the algorithm
        - return the item in that position (in order)
            - order can be increasing or decreaasing... read the Q carefully
            - if k=10, partition list of 50 items in increasing
                -10th smallest on the left, 40th biggest on the right
            
- How does it link to QuickSort?
    - quicksort does sorts, quickselect does only partition based on the wanted pivot position and the partitions are not in order
    - Partiotioning: both uses partitioning. O(N) to partition
    - Recursion: quicksort go both side (recursively), quickselect go one side (recursively)
- How does it link to binary search?
    - binary search = search for the element, if it exist
    - quickselect = search for the element based on the position (k-value)
    - both only recurse 1 side
