Suffix Trie
- How is it different from a Trie (prefix trie)
    - all the same
    - except of N words, you have M suffixes
        - M is the length of the word
        - M is the number of suffixes
        - M is also the length of the longest suffix
- Space if i have a word of size M
    - i want a suffix trie
    - i have M suffixes
    - longest suffix is M characters long
    - total is O(M+M-1+M-2+M-3) space.... O(M^2)
